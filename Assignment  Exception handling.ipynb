#### Que 4 : Why LookUpError class is used? Explain with an example KeyError and IndexError

Ans.  The 'LookupError' class is a base class for exceptions that occur when a lookup or indexing operation fails. 'LookupError' is used to catch general lookup-related exceptions and handle them in a uniform way.

KeyError and IndexError are the examples of LookUpError :


```python
# 1. KeyError :  This error is raised when a dictionary key is not found during a lookup operation or  when you try to access a non-existent key in a dictionary.
# Example :
try:
    my_dict = {1 : "rohan", 2 : "Data Science Masters",3 : "music" }
    my_dict["name"]
    
except KeyError as e :
    print("ERROR :  Key {} does not exist. ".format(e))
```

```python
# IndexError :   This error is raised when an index is out of range in a sequence or list. It occurs when you try to access an element at an invalid or non-existent index.

# Example:
try:
    l = [1,2,3,4,5,6,"name", True, 90, 10]
    l[15]    
except IndexError as e :
    print("ERROR : There's something wrong with the index that you gave amd that is ...",e)
    
```

#### Que 5 : Explain ImportError. What is ModuleNotFoundError?

Ans. 
ImportError : ImportError can be raised in various scenarios, such as when a module or package is not found, when there is a problem with the module's code, or when there are issues with the Python environment or configuration.

ModuleNotFoundError : It is subclass of ImportError, raised when a specific module cannot be found or imported.


```python
# Example 
try :
    import non_existent_module
    
except ModuleNotFoundError as e :
    print("ERROR :", e)
```
